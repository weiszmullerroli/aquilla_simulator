## choosed one of the most diverse state params - looks greak
## Aquilla settings: 6 atom ring, d0=7.5, t = 2us, omega=3.6, delta=(-36,36), 200 meauserment

```python
import json
import matplotlib.pyplot as plt
from collections import Counter

# Load JSON data
with open('superpos.json', 'r') as f:
    data = json.load(f)
measurements = data['measurements']

# Extract all postSequences as strings
all_states = [''.join(str(bit) for bit in m['shotResult']['postSequence']) for m in measurements]

# Count occurrences of each unique state
state_counts = Counter(all_states)
total_measurements = len(all_states)

# Sort states alphabetically for consistent plotting
states_sorted = sorted(state_counts.keys())
probabilities = [state_counts[s] / total_measurements for s in states_sorted]

# Plot
plt.figure(figsize=(12, 6))
plt.bar(states_sorted, probabilities, color='skyblue')
plt.xlabel('State')
plt.ylabel('Probability')
plt.title('Probabilities of All Observed States')
plt.xticks(rotation=90, fontsize=8)
plt.tight_layout()
plt.show()
```
