Simple Usage Example - Ground State Population Analysis

This script demonstrates the key functionality you need:
1. Plot ground state population evolution (like your attached image)
2. Single function that handles both quick tests and parameter scans
3. Focus on ground state |000000⟩ analysis after 4 μs

```python
# Aquila Quantum Simulation - Computational States Analysis
# Jupyter Notebook for testing and demonstrating the computational basis states analysis

# Cell 1: Import and Setup
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Import our custom Aquila simulation package
# (Make sure the aquila_quantum_sim.py file is in your working directory or PYTHONPATH)
from aquila_quantum_package import (
    AquilaSimulator, SimulationParams, ComputationalStateAnalyzer, 
    ParameterSweepAnalyzer, aquila_computational_analysis,
    quick_computational_test, computational_parameter_scan
)

print("Aquila Quantum Simulation - Computational States Analysis")
print("=" * 60)

# Cell 2: Test Single Simulation with Computational States Plot
print("1. SINGLE SIMULATION TEST")
print("-" * 30)

# Run a single simulation
result_single = quick_computational_test(
    d0=7.51,      # Distance parameter (μm)
    omega=8.0,    # Rabi frequency (MHz)  
    delta=30.0,   # Detuning (MHz)
    runtime=4.0,  # Simulation time (μs)
    show_plots=True
)

print(f"Ground state |000000⟩ probability: {result_single['basis_probabilities'].get('000000', 0):.4f}")
print(f"Total number of significant states: {len(result_single['basis_probabilities'])}")

# Cell 3: Manual Analysis of Computational States
print("\n2. DETAILED COMPUTATIONAL STATES ANALYSIS")
print("-" * 45)

# Create analyzer for detailed analysis
analyzer = ComputationalStateAnalyzer(n_atoms=6)

# Get the basis probabilities from our simulation
basis_probs = result_single['basis_probabilities']

# Categorize states by excitation level
categories = analyzer.categorize_states(basis_probs)

print("States by excitation level:")
for category, states in categories.items():
    n_excited = int(category.split('_')[0])
    total_prob = sum(states.values())
    print(f"  {n_excited} excited atoms: {len(states)} states, total prob = {total_prob:.4f}")
    
    # Show top 3 states in each category
    top_in_category = sorted(states.items(), key=lambda x: x[1], reverse=True)[:3]
    for state, prob in top_in_category:
        print(f"    |{state}⟩: {prob:.4f}")

# Cell 4: Create Custom Visualization
print("\n3. CUSTOM COMPUTATIONAL STATES VISUALIZATION")
print("-" * 50)

# Create a detailed bar plot
fig, ax = analyzer.create_state_barplot(
    basis_probs,
    title="Computational Basis States (d₀=7.51μm, Ω=8.0MHz, Δ=30.0MHz)",
    figsize=(16, 8),
    top_n=25,  # Show top 25 states
    show_excitation_groups=True
)
plt.show()

# Create excitation level summary
fig2, ax2 = analyzer.create_excitation_summary_plot(
    basis_probs,
    title="Total Probability by Excitation Level",
    figsize=(10, 6)
)
plt.show()

# Cell 5: Parameter Sweep for Computational States
print("\n4. PARAMETER SWEEP - COMPUTATIONAL STATES COMPARISON")
print("-" * 60)

# Define parameter ranges for sweep
d0_range = [7.0, 7.5, 8.0]           # Distance values (μm)
omega_range = [6.0, 8.0, 10.0]       # Rabi frequency values (MHz)
delta_range = [20.0, 30.0, 40.0]     # Detuning values (MHz)

print(f"Parameter ranges:")
print(f"  d0: {d0_range} μm")
print(f"  Omega: {omega_range} MHz") 
print(f"  Delta: {delta_range} MHz")
print(f"  Total combinations: {len(d0_range) * len(omega_range) * len(delta_range)}")

# Run parameter sweep
sweep_result = computational_parameter_scan(
    d0_range=d0_range,
    omega_range=omega_range,
    delta_range=delta_range,
    runtime=3.0,  # Shorter runtime for faster testing
    n_processes=4,  # Adjust based on your CPU
    show_plots=True
)

print(f"\nSweep completed! Analyzed {len(sweep_result['results'])} parameter combinations")

# Cell 6: Analyze Sweep Results
print("\n5. PARAMETER SWEEP ANALYSIS")
print("-" * 35)

# Print interesting findings
print("Most interesting parameter combinations:")
print(f"Lowest ground state retention: {sweep_result['min_ground_state'][0]} → {sweep_result['min_ground_state'][1]['ground_state_prob']:.4f}")
print(f"Highest excitation achieved: {sweep_result['max_excited_states'][0]} → {sweep_result['max_excited_states'][1]['max_excited_states']} atoms")
print(f"Most diverse states: {sweep_result['most_diverse_states'][0]} → {sweep_result['most_diverse_states'][1]['n_significant_states']} states")

# Cell 7: Compare Specific Parameter Sets
print("\n6. DETAILED COMPARISON OF SELECTED PARAMETERS")
print("-" * 55)

# Let's manually compare a few interesting cases
interesting_params = [
    (7.0, 8.0, 30.0),   # Low d0
    (8.0, 8.0, 30.0),   # High d0
    (7.5, 6.0, 30.0),   # Low omega
    (7.5, 10.0, 30.0),  # High omega
]

# Create a custom comparison plot
sweep_analyzer = ParameterSweepAnalyzer(d0_range, omega_range, delta_range)
sweep_analyzer.results = sweep_result['results']

# Filter for parameters we actually have
available_params = [p for p in interesting_params if p in sweep_result['results']]

if available_params:
    fig_comp, axes_comp = sweep_analyzer.create_comparison_plots(
        available_params,
        figsize=(20, 10),
        top_n=15
    )
    plt.suptitle("Computational States Comparison for Selected Parameters", fontsize=16, y=0.98)
    plt.show()

# Cell 8: Export Results and Summary Statistics
print("\n7. SUMMARY STATISTICS")
print("-" * 25)

# Calculate summary statistics across all parameter combinations
all_ground_probs = []
all_max_excitations = []
all_n_states = []

for key, analysis in sweep_result['interesting_analysis'].items():
    all_ground_probs.append(analysis['ground_state_prob'])
    all_max_excitations.append(analysis['max_excited_states'])
    all_n_states.append(analysis['n_significant_states'])

print(f"Ground state probability statistics:")
print(f"  Mean: {np.mean(all_ground_probs):.4f}")
print(f"  Std:  {np.std(all_ground_probs):.4f}")
print(f"  Min:  {np.min(all_ground_probs):.4f}")
print(f"  Max:  {np.max(all_ground_probs):.4f}")

print(f"\nMaximum excitation statistics:")
print(f"  Mean: {np.mean(all_max_excitations):.1f}")
print(f"  Max:  {np.max(all_max_excitations)}")

print(f"\nNumber of significant states statistics:")
print(f"  Mean: {np.mean(all_n_states):.1f}")
print(f"  Std:  {np.std(all_n_states):.1f}")
print(f"  Max:  {np.max(all_n_states)}")

# Cell 9: Create Final Summary Plot
print("\n8. FINAL SUMMARY VISUALIZATION")
print("-" * 40)

# Create a summary figure with multiple subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

# Plot 1: Ground state probabilities distribution
ax1.hist(all_ground_probs, bins=15, alpha=0.7, color='blue', edgecolor='black')
ax1.set_xlabel('Ground State Probability')
ax1.set_ylabel('Frequency')
ax1.set_title('Distribution of Ground State Probabilities')
ax1.grid(True, alpha=0.3)

# Plot 2: Max excitations distribution  
ax2.hist(all_max_excitations, bins=range(0, max(all_max_excitations)+2), 
         alpha=0.7, color='red', edgecolor='black')
ax2.set_xlabel('Maximum Excitation Level')
ax2.set_ylabel('Frequency')
ax2.set_title('Distribution of Maximum Excitation Levels')
ax2.grid(True, alpha=0.3)

# Plot 3: Number of significant states distribution
ax3.hist(all_n_states, bins=15, alpha=0.7, color='green', edgecolor='black')
ax3.set_xlabel('Number of Significant States')
ax3.set_ylabel('Frequency')
ax3.set_title('Distribution of Significant States Count')
ax3.grid(True, alpha=0.3)

# Plot 4: Correlation plot
ax4.scatter(all_ground_probs, all_n_states, alpha=0.6, color='purple')
ax4.set_xlabel('Ground State Probability')
ax4.set_ylabel('Number of Significant States')
ax4.set_title('Ground State vs State Diversity')
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nAnalysis complete! All plots and statistics generated.")
print("Files saved: computational_states_comparison.png and individual parameter plots")

# Cell 10: Advanced Analysis - Find Optimal Parameters
print("\n9. OPTIMAL PARAMETER IDENTIFICATION")
print("-" * 40)

# Define different optimization criteria
criteria = {
    'min_ground_state': lambda x: x['ground_state_prob'],  # Minimize ground state (maximize excitation)
    'max_diversity': lambda x: x['n_significant_states'],  # Maximize number of states
    'max_excitation': lambda x: x['max_excited_states'],   # Maximize excitation level
}

print("Optimal parameters for different criteria:")
for criterion_name, criterion_func in criteria.items():
    if criterion_name == 'min_ground_state':
        optimal = min(sweep_result['interesting_analysis'].items(), key=lambda x: criterion_func(x[1]))
        direction = "minimize"
    else:
        optimal = max(sweep_result['interesting_analysis'].items(), key=lambda x: criterion_func(x[1]))
        direction = "maximize"
    
    params, metrics = optimal
    d0, omega, delta = params
    print(f"\n{criterion_name} ({direction}):")
    print(f"  Parameters: d0={d0:.1f}μm, Ω={omega:.1f}MHz, Δ={delta:.1f}MHz")
    print(f"  Ground state prob: {metrics['ground_state_prob']:.4f}")
    print(f"  Max excitation: {metrics['max_excited_states']}")
    print(f"  N significant states: {metrics['n_significant_states']}")

print("\n" + "="*60)
print("ANALYSIS COMPLETE - Check generated plots and saved files!")
print("="*60)
```
